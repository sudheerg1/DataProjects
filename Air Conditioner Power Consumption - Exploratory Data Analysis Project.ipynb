### From the above two graphs we can conclude that power consumption seems to exhibit some kind of pattern. Lets delve deeper into data grouping by Hour, Day and Week, also Weekdays and Weekends 

```python
df['hour'] = df['timestamp'].dt.hour
```

```python
df.tail()
```

```python
g_hour = df.groupby('hour')
```

```python
hour_y = g_hour.sum()
hour_y
```

```python
plt.title('AC Power vs Clocktime')
plt.xlabel('Clocktime (in hrs)')
plt.ylabel('AC Power')

plt.plot(hour_y.index.tolist(),hour_y['AC 1'])
plt.plot(hour_y.index.tolist(),hour_y['AC 2'])
plt.plot(hour_y.index.tolist(),hour_y['AC 3'])
plt.plot(hour_y.index.tolist(),hour_y['AC 4'])
plt.plot(hour_y.index.tolist(),hour_y['AC 5'])
plt.plot(hour_y.index.tolist(),hour_y['AC 6'])
plt.plot(hour_y.index.tolist(),hour_y['AC 7'])
plt.plot(hour_y.index.tolist(),hour_y['AC 8'])
matplotlib.rcParams['figure.figsize'] = (12,8)
plt.show()
```

### 1) The above graph shows that consumption is minimal during day time (10 AM to 8 PM) on average. The peak usage time is between 11 PM and 5 AM. This conclusion is based on AC 1 data.

```python
hour_all = hour_y.sum(axis=1) #sum of consumption of all ACs
```

```python
hour_all
```

```python
plt.title('Cumulative AC Power vs Clocktime')
plt.xlabel('Clocktime (in hrs)')
plt.ylabel('Cumulative AC Power')

plt.plot(hour_all.index.tolist(),hour_all,color='blue')
```

### 1) The above plot between Power summed up for all ACs vs the Hourly Time shows that consumption raises from 8PM onwards with peak at 11 PM and falls back to nominal value at 10 AM.

## Correlation between Power Consumption and Temperature Outside

```python
temp['day_all'] = day_all 
```

```python
temp.head()
```

```python
temp.dtypes
```

```python
temp['max temp'] = temp['max temp'].astype(float)
temp['avg temp'] = temp['avg temp'].astype(float)
temp['min temp'] = temp['min temp'].astype(float)
```

```python
corr_matrix1 = temp.corr(method='pearson') #temp.corr(method='spearman') #temp.corr(method='kendall')
corr_matrix1
```

```python
corr_pairs = corr_matrix1.unstack() 
corr_pairs['day_all'] #Correlation of Daily Temperature with Daily Power Consumption
```

```python
corr_matrix = temp.corr(method='pearson')
sns.heatmap(corr_matrix,annot=True, cmap='gray')
plt.show()
matplotlib.rcParams['figure.figsize'] = (8,8)
```

### The correlation coefficient between power consumption and outside temperature is around "0.5". Since it is less than 0.8 we can consider the relation is not significant. 

### Here, temperature relationship has been check on daily basis so cannot comment on minute wise relationship between power consumption and temperature (non availability of data)

## The graph is highly non linear. The data is not seem to be exhibiting any trend. However a trendline can be constructed assuming the population data is linear.

```python
m,c = np.polyfit(temp.index.tolist(),temp['day_all'],1)

# y = mx + c
```

```python
print(m,c)
```

```python
plt.title('Daily AC Power Cumulative vs Time')
plt.xlabel('Time (in Days)')
plt.ylabel('Daily AC Power Cumulative')

plt.plot(temp.index,temp['day_all'],color='blue')
plt.plot(temp.index,m*temp.index+c,color='red')
plt.show()
```
