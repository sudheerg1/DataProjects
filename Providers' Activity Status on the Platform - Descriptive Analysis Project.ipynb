```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data Engineer Assignment.csv")
```

```python
df.head()
```

```python
df['event_time']=pd.to_datetime(df['event_time'])
```

```python
df.head()
```

```python
df.shape
```

```python
df[df.duplicated()].shape #check for duplicates 
```

```python
df.drop_duplicates(inplace=True) #dropping all the duplicate rows
df[df.duplicated()].shape
```

```python
df.shape
```

```python
df.isnull().values.any() #check for any null values 
```

```python
df['offline_status'] = df['detail'].str.contains("False")
df['offline_status'].replace(True,'offline', inplace=True)
df['offline_status'].replace(False, np.nan, inplace=True)
```

```python
df['online_status'] = (df['detail'].str.contains("True")) | (df['source'].str.contains("Action on Job"))
df['online_status'].replace(True,'online', inplace=True)
df['online_status'].replace(False, np.nan, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(axis='index', how='all', subset=['offline_status', 'online_status'], inplace=True)
df.head()
```

```python
df.shape
```

```python
df['offline_status'].replace('offline', True, inplace=True)
df['offline_status'].replace(np.nan, False, inplace=True) 
df['online_status'].replace('online', True, inplace=True)
df['online_status'].replace(np.nan, False, inplace=True)
df.head()
```

```python
df.isnull().values.any() 
```

```python
df['status'] = (df['online_status'] & ~df['offline_status']) | (df['online_status'] & df['offline_status']) #Boolean Expression derived from Boolean Algebra
df.head()
```

```python
df['status'].replace(True,'online', inplace=True)
df['status'].replace(False,'offline', inplace=True)
df.head()
```

```python
df.drop(['detail','source','offline_status','online_status'], axis=1, inplace=True)
df.head()
```

```python
df['status'].value_counts()   #1.Num of online events and 2.Num of offline events
```

```python
df.drop('hour', axis=1, inplace=True)
df.head()
```

```python
provider_list = df['provider_id'].unique().tolist()
```

```python
#print(provider_list)
```

```python
time_init = pd.Timestamp('2017-09-01 19:00:00')
time_init
```

```python
time_delta = pd.Timedelta('1 day')
time_delta
```

```python
time_list = []
time = time_init
for i in range(30):
    time_list.append(time)
    time += time_delta
```

```python
#time_list
```

```python
list_of_tuples = []

for provider in provider_list:
    for time in time_list:
        list_of_tuples.append((provider,time,'offline'))
```

```python
df_EOD_offline = pd.DataFrame(list_of_tuples, columns = ['provider_id','event_time','status'])

#Reset to offline at EOD - For each provider for each date need added a row with timestamp "2017-09-xx 19:00:00" - 'provider_id' - 'offline'
```

```python
df_total = pd.concat([df,df_EOD_offline], ignore_index=True)
```

```python
df_total.tail()
```

```python
df_total = df_total.set_index('event_time')
df_total.head()
```

```python
df_total['time'] = df_total.index
df_total.head()
```

```python
df_total.sort_values(['provider_id','time'], inplace=True)
```

```python
df_total.head()
```

```python
df_total['date'] = df_total['time'].map(pd.Timestamp.date)
df_total['hour startime'] = df_total['time'].dt.hour
df_total['hour endtime'] = df_total['hour startime'] + 1
```

```python
df_total.head()
```

```python
df_total['delta'] = df_total['time']-df_total['time'].shift()
```

```python
df_total.head()
```

```python
df_total['delta']=pd.to_timedelta(df_total['delta'], errors='coerce')
```

```python
df_total.isnull().values.any()
```

```python
df_total.shape
```

```python
df_total.dropna(inplace=True)
```

```python
df_total.shape
```

```python
df_total['delta1'] = df_total['delta'].mask(df_total['status'].shift(-1)=='offline',0)  
#time where the provider is online
```

```python
df_total.head()
```

```python
#df_total.dtypes
```

```python
df_total['delta1'] = pd.to_timedelta(df_total['delta1']) #conversion from object datatype to timedelta64[ns]
```

```python
df_total['delta1'] = df_total['delta1'].dt.total_seconds() #conversion of delta1 to seconds online
```

```python
df_total.head(10)
```

```python
df_total[df_total['delta1']>0].head()
```

```python
df_new = df_total[df_total['hour startime']!=19]
```

```python
g = df_new.groupby(['provider_id','date','hour startime','hour endtime'])
```

```python
final_df = g.sum()
```
